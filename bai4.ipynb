```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
gdf_population = gpd.read_file("https://raw.githubusercontent.com/TungTh/tungth.github.io/master/data/vn-provinces.json")
col_name_pop = 'Name' if 'Name' in gdf_population.columns else gdf_population.columns[0]
danh_sach_tinh_pop = gdf_population[col_name_pop].tolist()
np.random.seed(43)
dan_so_gia_lap = np.random.randint(500000, 10000000, size=len(danh_sach_tinh_pop))
df_dan_so = pd.DataFrame({'Ten_Khu_Vuc': danh_sach_tinh_pop, 'Dan_So': dan_so_gia_lap})
merged_gdf_population = gdf_population.merge(df_dan_so, left_on=col_name_pop, right_on='Ten_Khu_Vuc')
fig, ax = plt.subplots(1, 1, figsize=(12, 14))
merged_gdf_population.plot(
    column='Dan_So',
    cmap='YlGnBu',
    linewidth=0.5,
    ax=ax,
    edgecolor='gray',
    legend=True,
    legend_kwds={'label': "Dân số (Người)", 'orientation': "horizontal", 'shrink': 0.7}
)
ax.set_title("BẢN ĐỒ CHOROPLETH: DÂN SỐ THEO TỈNH THÀNH", fontdict={'fontsize': 16, 'fontweight': 'bold'})
ax.axis('off')
plt.tight_layout()
plt.show()
```
