```python
#Bài 23.2
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx
m = folium.Map(location=[10.775, 106.700], zoom_start=13)
warehouse = (10.7769, 106.682)
address =[
    "Van Hanh Mall,Ho Chi Minh City",
    "Giga Mall,Ho Chi Minh City",
    "UEH,Ho Chi Minh City",
    "Dam Sen,Ho Chi Minh City",
    "Benh Vien Cho Ray,Ho Chi Minh City",
    "Landmark 81,Ho Chi Minh City",
    "Dinh Doc Lap,Ho Chi Minh City",
    "Bitexco Financial Tower,Ho Chi Minh City",
    "San bay quoc te Tan San Nhat,Ho Chi Minh City",
    "Ben Thanh Market,Ho Chi Minh City",
    "Sân vận động Thống Nhất,Ho Chi Minh City",]
x=[]
y=[]
geolocator = Nominatim(user_agent="routing-example", timeout=10) # Added timeout
folium.Marker(warehouse,popup="Trung Tam",icon=folium.Icon(color='red')).add_to(m)
G = ox.graph_from_point(warehouse, dist=10000, network_type='drive')
orig_node = ox.distance.nearest_nodes(G, warehouse[1], warehouse[0])

for addr_str in address:
  location = geolocator.geocode(addr_str)
  if location:
    x.append(location.latitude)
    y.append(location.longitude)
    folium.Marker([location.latitude, location.longitude], popup=addr_str, icon=folium.Icon(color='blue')).add_to(m)
    dest_node = ox.distance.nearest_nodes(G, location.longitude, location.latitude)
    try:
        route = nx.shortest_path(G, orig_node, dest_node, weight='length')
        route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]
        folium.PolyLine(locations=route_coords, color='green', weight=3, opacity=0.8).add_to(m)
        route_length_m = nx.path_weight(G, route, weight='length')
        route_length_km = route_length_m / 1000
        print(f"Tuyến đường từ nhà kho đến {addr_str}: {route_length_km:.2f} km")
    except nx.NetworkXNoPath:
        print(f"Không tìm thấy đường đi từ nhà kho đến {addr_str}")
    except Exception as e:
        print(f"Lỗi khi tính tuyến đường đến {addr_str}: {e}")

    print(f"Tên: {location.address}")
    print(f"Tọa độ: {location.latitude}, {location.longitude}")
  else:
    print(f"Không tìm thấy tọa độ cho địa chỉ: {addr_str}") # Print message if address is not found
m
```
